```python
import pandas as pd

# Assuming you have loaded your data into a DataFrame called df

# List of columns containing innings values
innings_columns = ['Inns.0_x', 'Inns.1_x', 'Inns.2_x', 'Inns.3_x', 'Inns.4_x', 'Inns.5_x',
                   'Inns.0_y', 'Inns.1_y', 'Inns.2_y', 'Inns.3_y', 'Inns.4_y', 'Inns.5_y']

# Replace non-numeric values or missing values with 0
india[innings_columns] = india[innings_columns].apply(pd.to_numeric, errors='coerce').fillna(0)

# Create the 'innings' column by summing all innings values
india['innings'] = india[innings_columns].sum(axis=1)
```

```python
# List of columns to drop
columns_to_drop = ['Inns.0_x', 'Inns.1_x', 'Inns.2_x', 'Inns.3_x', 'Inns.4_x', 'Inns.5_x',
                   'Inns.0_y', 'Inns.1_y', 'Inns.2_y', 'Inns.3_y', 'Inns.4_y', 'Inns.5_y']

# Drop the columns
india = india.drop(columns=columns_to_drop)
```

**RUNS**

```python
import pandas as pd

# Assuming you have loaded your data into a DataFrame called india

# List of columns containing runs values
runs_columns = ['Runs.0_x', 'Runs.1_x', 'Runs.2_x', 'Runs.3_x', 'Runs.4_x', 'Runs.5_x',
                'Runs.0_y', 'Runs.1_y', 'Runs.2_y', 'Runs.3_y', 'Runs.4_y', 'Runs.5_y']

# Replace non-numeric values or missing values with 0
india[runs_columns] = india[runs_columns].apply(pd.to_numeric, errors='coerce').fillna(0)

# Create the 'runs' column by summing all runs values
india['runs'] = india[runs_columns].sum(axis=1)
```

```python

# List of columns to drop
columns_to_drop = ['Runs.0_x', 'Runs.1_x', 'Runs.2_x', 'Runs.3_x', 'Runs.4_x', 'Runs.5_x',
                'Runs.0_y', 'Runs.1_y', 'Runs.2_y', 'Runs.3_y', 'Runs.4_y', 'Runs.5_y']

# Drop the columns
india = india.drop(columns=columns_to_drop)
```

```python
wic_columns = ['Wkts.0', 'Wkts.1', 'Wkts.2', 'Wkts.3', 'Wkts.4', 'Wkts.5']

# Replace non-numeric values or missing values with 0
india[wic_columns] = india[wic_columns].apply(pd.to_numeric, errors='coerce').fillna(0)

# Create the 'wicket' column by summing all wickets values
india['wicket'] = india[wic_columns].sum(axis=1)
```
